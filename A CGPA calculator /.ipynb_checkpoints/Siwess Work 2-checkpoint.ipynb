```python
import numpy as np
import pandas as pd
from numpy import arange
from numpy.random import randint
from numpy import array
from pandas import DataFrame as df
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
q_t= randint(0,40,160).reshape(20,8)
q_e= randint(1,60,160).reshape(20,8)
r = arange(75,95,1)
t = 'MTH301 CPE310 CPE316 CPE314 CSC302 CSC306 CSC308 EEE301'.split()
```

```python
i = 79
while i < 100: 
    print('CSC/2017/{}'.format(i))
    i = i+1
```

```python
i = arange(79,99)
b = np.array(['CSC/2017/' + str(i) for i: in range(79,99)])
b
```

```python
test_scores = df(q_t,b,t)
test_scores
```

```python
#test_scores['MTH301'].apply(lambda(x :x in ab(s)))
mth_absent = test_scores[test_scores['MTH301'] == 0]['MTH301']
cpe310_absent = test_scores[test_scores['CPE310'] == 0]['CPE310']
cpe314_absent =test_scores[test_scores['CPE314'] == 0]['CPE314']
cpe316_absent=test_scores[test_scores['CPE316'] == 0]['CPE316']
csc302_absent=test_scores[test_scores['CSC302'] == 0]['CSC302']
csc306_absent=test_scores[test_scores['CSC306'] == 0]['CSC306']
csc308_absent=test_scores[test_scores['CSC308'] == 0]['CSC308']
eee301_absent=test_scores[test_scores['EEE301'] == 0]['EEE301']
absentees = pd.Series(data = [mth_absent,cpe310_absent,cpe314_absent,cpe316_absent,csc302_absent,csc306_absent,csc308_absent,eee301_absent], index = ['Absent in Mth 301','Absent in CPE 310','Absent in CPE 314','Absent in CPE 316','Absent in CSC 302','Absent in CSC 306','Absent in CSC 308','Absent in EEE 301'], name = 'Absent')
absentees
```

```python
exam_scores = df(q_e,b,t)
exam_scores
```

```python
Total = test_scores + exam_scores
Total
```

```python
#Total['MTH301'].plot.hist()
student,course = plt.subplots(2,4, figsize=(18,10))
sns.histplot(Total['MTH301'],ax = course[0,0],bins =7, kde = False)
sns.histplot(Total['CPE310'], ax = course[0,1], kde = False)
sns.histplot(Total['CPE314'], ax = course[0,2],kde = False)
sns.histplot(Total['CPE316'], ax = course[0,3], kde = False)
sns.histplot(Total['CSC302'], ax = course[1,0], kde = False)
sns.histplot(Total['CSC306'], ax = course[1,1],kde = False)
sns.histplot(Total['CSC308'], ax = course[1,2],kde = False)
sns.histplot(Total['EEE301'], ax = course[1,3], kde = False)
```

```python
#Grading
def grade(x):
    if x > 70:
        return 'A'
    elif 59<x<70:
        return 'B'
    elif 49<x<60:
        return 'C'
    elif 44<x<50:
        return 'D'
    elif 39<x<45:
        return 'E'
    else:
        return 'F'
```

```python
t1 = Total.copy()
for x in t1.columns:
    t1[x] = t1[x].apply(lambda x: grade(x))
```

```python
t1
```

```python
#Points Calculator
def points(p):
    if p == 'A':
        return 5.0
    elif p == 'B':
        return 4.0
    elif p == 'C':
        return 3.0
    elif p == 'D':
        return 2.0
    elif p == 'E':
        return 1.0
    else:
        return 0.0
```

```python
t2 = t1.copy()
for m in t2.columns:
    t2[m] = t2[m].apply(lambda p:points(p))
```

```python
units = array([5,3,2,1,3,2,3,3])
points = t2 * units
Student_GP = points.sum(axis=1)/sum(units)
Student_GP
```
